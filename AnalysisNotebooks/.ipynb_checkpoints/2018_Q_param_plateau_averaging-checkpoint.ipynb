```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
#from scipy.optimize import curve_fit
from IPython.core.pylabtools import figsize
figsize(10, 7)

# Enter no. of re-runs
runs=20

# Combine data files for multiple runs
for f in glob.glob("run*.dat"):
         os.system("cat "+f+" >> all_data.dat")

# Plot multiple systems together
exptl_data= np.genfromtxt('expt_CZTSe_Q_vs_T.csv', delimiter = ',' ) # Exptl reference
file1= np.genfromtxt('run1.dat', delimiter = ' ') # Single run to compare to
all_data = np.genfromtxt('all_data.dat', delimiter = ' ') # Combined data from all runs

plt.xlabel('Simulation temperature ($^{\circ}$C)')
plt.ylabel('Order parameter Q')
plt.plot(exptl_data[:,0]+70, exptl_data[:,1], label ='Experiment')
plt.scatter(all_data[:,0]-273.15, abs(all_data[:,1]), marker="x", color="orange",  s=10, label='Monte Carlo')


# Add polynomial line of best fit to all_data
x = all_data[:,0]-273.15
y = abs(all_data[:,1])

z = np.polyfit(x, y, 20) # deg of polynomial just chosen to give best shape to curve
f = np.poly1d(z)
x_new = np.linspace(x[0], x[-1], 50)
y_new = f(x_new)
plt.plot(x,y,'o', x_new, y_new, color="orange")

plt.legend(loc='upper right', frameon=False)
plt.rcParams.update({'font.size': 20})

plt.xlim(0,700)
plt.ylim(-0.1,1.0)

plt.savefig("plateau+expt.png")
plt.show()
```
