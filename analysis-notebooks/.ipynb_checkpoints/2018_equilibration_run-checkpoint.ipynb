# Eris equilibration run

This notebook contains instructions and scripts for performing an equilibration run to determine a suitable number for `MCEqmSteps` in eris.cfg, i.e. the number of Monte carlo steps to perform in a production run before using the data for final analysis of an equilibrated system. 

To check for the number of MC steps required to reach equilibrium, we calculate the variance of the distribution of Sn ions over a number of MC steps and looks for the point at which the variance reaches a steady value. Note that it may be necessary to alter the parameter `MCS_av` to change the number of MC steps included in each variance calculation to reduce the amplitude of fluctuations in the plot.

This needs to be performed for different system sizes and across the temperature range. Larger systems require a larger number of MC steps to equilibrate and there can also be a 'criticial slowing down' at temperatures close to a phase transition.

This notebook will perform the equilibration steps check across the temperature range for each system, but the analysis will need to be repeated if simulating a larger system.

## Eris simulation flags
Use the eris.cfg flag: EquilibrationChecks=true
(note that the EquilibrationChecksTestCutOff is not needed for a simple equilibration run, details of this check are included in the SI of the publication cited on the repo README)

The EquilibrationChecks flag calculates the on-site electrostatic potentials of Sn ions with a finite cut off radius in the 'analysis_midpoint' function in eris-main.c. The Sn potentials are recalculated after each step of the outer Monte Carlo loop, referred to as 'MCMegaSteps' in Eris.

## Eris Outputs
This simulation flag outputs Sn_potentials_T_TEMP.dat files into a directory called  ‘equilibration_check_Sn_potentials’. These can then be plotted using the following python scripts.

**UPDATE:** new version of code instead writes e.g. 'Sn_Pot_T_1000_MCS_00031.dat' to same directory, but _initial still have same name but now also in same directory as general outputs (below scripts have now been updated for this and MCS loop range modified to match new outputs)

### Equilbration run for a single temperature
Set the desired simulation temperature in the following python script, the number MCMegaSteps was set to in eris.cfg and also the number of data points you wish to calculate the variance for.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------------------------------------------------------------------------------------------------------------
# USER INPUTS: Choose temperature to plot variance as a function of MCS for + MCMegaSteps (MCS_max) as defined in eris.cfg and number of MC steps to average over for each data point in plot
# -----------------------------------------------------------------------------------------------------------------------------------

temp=550
MCS_max=100
MCS_av=5

# -----------------------------------------------------------------------------------------------------------------------------------

# Initializing empty variance file
variance_file = []


# Calculating variance of initial lattice
file="Sn_potentials_Temp_"+str(temp).zfill(4)+"_initial.dat"
data = np.loadtxt(file)
variance = np.var(data)
variance_file.append(variance)


# Calculating variance averaged over set number of configurations
# Defining inner and out MC loops for averaging
MC_inner_max=MCS_av
MC_outer_max=MCS_max-MCS_av
outer_loop=range(1,MC_outer_max+1,MCS_av)
inner_loop=range(0,MC_inner_max,1)

# Outer MC step loop
for j in outer_loop:

    # Re-initialise empty file for list of on-site potentials for several configs to average over
    pot_to_av=[]

    # Inner MC step loop
    for i in inner_loop:
        k = i+j
        file ="Sn_Pot_T_"+str(temp).zfill(4)+"_MCS_"+str(k).zfill(5)+".dat"
        pot = np.loadtxt(file)
        pot_to_av.append(pot)

    # Take variance of several configs after inner MC loop
    variance = np.var(pot_to_av)
    variance_file.append(variance)



# Defining list for MC steps for x-axis
MCS=range(1, MCS_max+1, MCS_av)

#print(MCS, variance_file)

plt.plot(MCS, variance_file)
plt.xlabel('MC Mega Step ('+str(MCS_av)+' per data point)')
plt.ylabel('Variance in Distribution of Electrostatic Potentials of Sn Ions')
plt.title('Temperature: '+ str(temp) +'K')

plt.xticks(fontsize=14)

plt.show()
```

### Or equilbration run for a across a temperature range
Set the desired simulation temperature range in the following python script, the number MCMegaSteps was set to in eris.cfg and also the number of data points you wish to calculate the variance for.

Note that this last plot may take a whiel to plot!

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from IPython.core.pylabtools import figsize
figsize(9, 7)

# -----------------------------------------------------------------------------------------------------------------------------------
# USER INPUTS: Choose temperature to plot variance as a function of MCS for + MCMegaSteps (MCS_max) as defined in eris.cfg and number of MC steps to average over for each data point in plot
# -----------------------------------------------------------------------------------------------------------------------------------

temp_min=50
temp_max=950
temp_step=100
MCS_max=100
MCS_av=5

# initialise color
temp_range=(temp_max-temp_min)/temp_step
color_shift=np.linspace(0.1, 0.9, 10)
c_index=0

# -----------------------------------------------------------------------------------------------------------------------------------


for color_index, temp in zip(color_shift, range(temp_min, temp_max, temp_step)):

    # Initializing empty variance file (new for each T)
    variance_file = []


    # Calculating variance of initial lattice
    file="Sn_potentials_Temp_"+str(temp).zfill(4)+"_initial.dat"
    data = np.loadtxt(file)
    variance = np.var(data)
    variance_file.append(variance)
  

    # Calculating variance averaged over set number of configurations
    # Defining inner and out MC loops for averaging
    MC_inner_max=MCS_av
    MC_outer_max=MCS_max-MCS_av
    outer_loop=range(1,MC_outer_max+1,MCS_av)
    inner_loop=range(0,MC_inner_max,1)

    # Outer MC step loop
    for j in outer_loop:

        # Re-initialise empty file for list of on-site potentials for several configs to average over
        pot_to_av=[]

        # Inner MC step loop
        for i in inner_loop:
            k = i+j
            file ="Sn_Pot_T_"+str(temp).zfill(4)+"_MCS_"+str(k).zfill(5)+".dat"
            pot = np.loadtxt(file)
            pot_to_av.append(pot)

        # Take variance of several configs after inner MC loop
        variance = np.var(pot_to_av)
        variance_file.append(variance)



    # Defining list for MC steps for x-axis
    MCS=range(1, MCS_max+1, MCS_av)

    #  print(MCS, variance_file)

    plt.plot(MCS, variance_file, color=plt.cm.bone(color_index), lw=2, label=str(temp)+' K')
    axis_font = {'fontname':'Arial', 'size':'18'}
    plt.xlabel('Mega Monte Carlo step ('+str(MCS_av)+' per data point)', **axis_font)
    plt.ylabel('Variance of Sn ion potentials', **axis_font)
    plt.legend(fontsize='x-large', frameon=False, bbox_to_anchor=(1.18, 0.8))
    plt.xticks(**axis_font)
    plt.yticks(**axis_font)
    plt.xlim([0,60])

plt.savefig('equil_all2.png')
plt.show()
```
